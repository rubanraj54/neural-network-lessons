# 1. Neurons

Neurons are simple individual components which takes an input and process it using some function and produce output.

This name is derived from human brain.

Our brain consists of 100 billion (100,000,000,000) neurons.. Its really a huge number.

If you want to do some actions, some neurons will be triggered and communicated with each other to accomplish that action.

Same thing is happening in the neural network, there are many neurons in the network to do some tasks.

# Activation Function

Each neuron have activation function.

Activation functions decides whether the current neuron is activated or not for the given input.

**How it decides? **

Lets consider, the threshold of the activation function is 1 then if the output exceeds the threshold value then the neuron will be activated else not.

Based on the activation output, we can predict the input. We'll see practical demo in upcoming lessons.

```python
sign_coordinates = np.array([[-2,-1],
                            [0,-1],
                            [0,0],
                           [0,1],
                           [2,1]])
```

```python
plt.figure(figsize=(10,5))
plt.grid(True)
plt.title("sign function")
plt.xlabel('x-coordinate')
plt.ylabel('y-coordinate')

plt.plot(sign_coordinates[:,0:1],sign_coordinates[:,1:2],'ro-')
```

```python
def sign(x):
    return -1 if (x < 0) else 1
```

```python
sign(0)
```
